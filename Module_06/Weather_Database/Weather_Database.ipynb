```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=200)

coordinates = zip(lats, lngs)
```

```python
from citipy import citipy

cities =[]
for lats,lngs in coordinates:
    city = citipy.nearest_city(lats,lngs)
    cities.append(city.city_name)
    
# cities
```

```python
# Import the requests library.
import requests

# Import the API key.
from config1 import weather_api_key

# Starting URL for Weather Map API Call.
baseurl = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(baseurl)
```

```python
# Create an endpoint URL for a city.
# Import the datetime module from the datetime library.
from datetime import datetime
import pandas as pd
# for city in cities:
#     city_url = baseurl + "&q=" + city
#     city_weather = requests.get(city_url)
#     city_weather_json = city_weather.json()
#     if(city_weather_json["cod"] == 200):
# #         Latitude and longitude
# #         Maximum temperature
# #         Percent humidity
# #         Percent cloudiness
# #         Wind speed
# #         Weather description (for example, clouds, fog, light rain, clear sky)
#         lat_lng = str(city_weather_json["coord"]["lat"]) + "," + str(city_weather_json["coord"]["lon"])
#         maximum_temperature =  str(city_weather_json["main"]["temp_min"]) + ","+ str(city_weather_json["main"]["temp_max"])
#         print(maximum_temperature)
#     else:
#         print("weather not found")

city_data =[]
        
for city in cities:
    try:
        # Parse the JSON and retrieve data.
        city_url = baseurl + "&q=" + city
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_country = city_weather["sys"]["country"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        current_description = city_weather["weather"][0]["description"]
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list.
        city_data.append({"City": city,
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description":current_description,
                          "Date": city_date})
        print("city ", city , " parsed.")
    # If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass


weather_df = pd.DataFrame(city_data)

weather_df


```
