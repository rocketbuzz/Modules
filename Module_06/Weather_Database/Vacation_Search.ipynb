```python
# Dependencies and Setup
import pandas as pd
import requests
# import gmaps

# Configure gmaps API key
# gmaps.configure(api_key=g_key)
```

```python
# 1. Import the WeatherPy_database.csv file. 
city_data_df = pd.read_csv("WeatherPy_database.csv")
city_data_df.head()
```

```python
# 2. Prompt the user to enter minimum and maximum temperature criteria 

min_temp = float(input('Enter minimum temperature:'))
max_temp = float(input('Enter maximum temperature:'))
```

```python
# 3. Filter the city_data_df DataFrame using the input statements to create a new DataFrame using the loc method.

mask =(city_data_df["Max Temp"] >= min_temp ) & (city_data_df["Max Temp"] <= max_temp)
city_data_filtered_df =city_data_df.loc[mask]
city_data_filtered_df

```

```python
# 4a. Determine if there are any empty rows.
city_data_filtered_df.isnull()
# city_data_filtered_df.dropna()
```

```python
# 4b. Drop any empty rows and create a new DataFrame that doesn’t have empty rows.

clean_df = city_data_filtered_df.dropna()

clean_df.head(5)
```

```python
# 5a. Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = clean_df[["City", "Country", "Lat", "Lng", "Max Temp", "Current Description"]].copy()
hotel_df

# 5b. Create a new column "Hotel Name"
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Import API key
from config1 import g_key

# 6a. Set parameters to search for hotels with 5000 meters.
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key
}

# 6b. Iterate through the hotel DataFrame.
for ind in hotel_df.index:
    # 6c. Get latitude and longitude from DataFrame
    location =str(hotel_df["Lat"][ind]) + ","+ str(hotel_df["Lng"][ind])

    # 6d. Set up the base URL for the Google Directions API to get JSON data.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    params["location"] = location

    # 6e. Make request and retrieve the JSON data from the search. 
    google_response = requests.get(base_url, params=params).json()
    
    try:
        # 6f. Get the first hotel from the results and store the name, if a hotel isn't found skip the city.
        hotel_df.loc[ind, "Hotel Name"] = google_response["results"][0]["name"]
#         If an error is experienced, skip the city.
    except:
        print("Hotel not found. Skipping...")
        pass


        
hotel_df
        
```

```python
# 7. Drop the rows where there is no Hotel Name.

valid_hotelname_mask =(hotel_df["Hotel Name"] != '')
clean_hotel_df = hotel_df.loc[valid_hotelname_mask]
clean_hotel_df

# hotel_df["Hotel Name"][72]

# import numpy as np
# hotel_df["Hotel Name"].replace('',np.nan, inplace=true)
# clean_hotel_df = hotel_df.dropna()
# clean_hotel_df

# hotel_df
```
